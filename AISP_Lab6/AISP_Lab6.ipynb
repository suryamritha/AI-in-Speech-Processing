A1. Take a portion of your recorded signal which represents a vowel sound. Perform FFT on the signal snippet and observe the amplitude spectrum. Repeat the same for a few vowel sounds.

A2. Repeat the A1 for a consonant sound. Perform the same for a few consonant sounds.


```python
import numpy as np
import matplotlib.pyplot as plt
import librosa
import librosa.display

def load_audio(file_path):
    signal, sr = librosa.load(file_path, sr=None)
    return signal, sr

def plot_amplitude_spectrum(signal, sr, title):
    # Perform FFT
    fft_result = np.fft.fft(signal)
    magnitude_spectrum = np.abs(fft_result)[:len(fft_result)//2]  # Take only positive frequencies

    # Frequency axis
    freq_axis = np.fft.fftfreq(len(signal), d=1/sr)[:len(fft_result)//2]

    # Plotting
    plt.figure(figsize=(10, 4))
    plt.plot(freq_axis, magnitude_spectrum)
    plt.title(title)
    plt.xlabel('Frequency (Hz)')
    plt.ylabel('Amplitude')
    plt.grid(True)
    plt.show()

def generate_spectrogram(signal, sr, title):
    plt.figure(figsize=(10, 4))
    spectrogram = librosa.display.specshow(librosa.amplitude_to_db(librosa.stft(signal), ref=np.max),
                                            sr=sr, x_axis='time', y_axis='log')
    plt.colorbar(format='%+2.0f dB')
    plt.title(title)
    plt.show()

# A1: Vowel sounds
vowel_files = ['e.m4a', 'i.m4a', 'o.m4a']
for file in vowel_files:
    signal, sr = load_audio(file)
    plot_amplitude_spectrum(signal, sr, f'Amplitude Spectrum of Vowel Sound ({file})')
    generate_spectrogram(signal, sr, f'Spectrogram of Vowel Sound ({file})')

# A2: Consonant sounds
consonant_files = ['j.m4a', 'k.m4a', 'm.m4a']
for file in consonant_files:
    signal, sr = load_audio(file)
    plot_amplitude_spectrum(signal, sr, f'Amplitude Spectrum of Consonant Sound ({file})')
    generate_spectrogram(signal, sr, f'Spectrogram of Consonant Sound ({file})')


```

A3. Repeat A2 for few slices of silence & non-voiced portions of the recorded speech signal.
